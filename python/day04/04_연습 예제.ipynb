```python
# 다음과 같이 리스트가 있을 때 판다스를 이용하여 표로 변환하세요

import pandas as pd

names = ["홍길동", "둘리", "고길동"]
kor_scores = [100, 90, 100]
eng_scores = [90, 10, 0]

# 이름 국어 영어
# 홍길동 100 90
# 둘리 90 10
# 고길동 100 0

df = pd.DataFrame ({
  '이름' : names,
  '국어' : kor_scores,
  '영어' : eng_scores
})

print(df)
```

```python
# 국어 성적 평균을 조회

print(f"국어 성적 평균 : {df['국어'].mean()}")
```

```python
# 평균 열을 추가(기본값은 0으로)

df['평균'] = 0
print(df)
```

```python
# 평균 열의 값을 국어와 영어의 평균으로 계산해서 넣어줌

df['평균'] = (df['국어'] + df['영어']) / 2
print(df)
```

```python
# 평균2 열을 추가 : 국어와 영어의 평균으로 계산해서 넣어줌

df['평균2'] = df[['국어','영어']].mean(axis=1)
print(df)
```

```python
# 국어 최고점을 조회

kor_max = df['국어'].max()


print(kor_max)

# 국어 최고점인 이름을 조회
print(df.loc[df['국어'] == kor_max,'이름'].values)
print(df.loc[df['국어'] == kor_max, '이름'].tolist())
print(df.loc[df['국어'] == df['국어'].max(), '이름'].tolist())
```
