```python
과일가게 = {"바나나", "사과", "딸기"}
for 과일 in 과일가게 : 
    print(f"과일가게에는 {과일}(이/가) 있습니다")

# for문 : min부터 max 까지 step씩 증가
min, max, step = 1, 10, 3
for i in range (min, max + 1, step):
    print(i)
```

```python
# 리스트 컴프리헨션 예제
# 1이상 10이하의 정수를 리스트에 넣어 확인하기 
list = [ i  for i in range (1, 10 + 1) if i % 2 == 0]
print(list)
```

```python
# 1이상 10이하의 정수들 중 짝수의 제곱을 리스트에 넣어 확인하기

list = [ i ** 2 for i in range ( 1, 10 + 1 ) if i % 2 == 0]
print(list)
```

```python
# 구구단 2단을 while문으로 출력하는 예제
num = 2
i = 1
while i <= 9 :
    print(f"{num} x {i} = { num * i }")
    i += 1
```

```python
# num이 소수인지 아닌지 판별하는 코드를 작성하세요
# 단, num은 2이상의 정수라고 가정
# input(문구) : 문구 출력 후 콘솔에서 입력 받은 값을 문자열로 가져옴

num = int(input("2이상의 정수를 입력하세요."))
count = 0

for i in range(1, num + 1,  1) : 
    if num % i == 0 :
        count += 1
if count == 2 :
    print(f" {num}은 소수")
else : 
    print(f" {num}은 소수가 아닙니다")
```
