```python
# 두 정수의 합을 반환하는 함수 선언
def sum(num1, num2) : 
  return num1 + num2 

# 함수 호출
print(sum(1,2))
```

```python
nums = [ 1, 3, 5, 7, 9, 11]

# 리스트에서 가장 큰 수를 반환하는 함수를 선언
def find_max(nums) :
  max_num = nums[0]

  for n in nums :
    if n > max_num :
      max_num = n

  return max_num


# 함수를 호출해서 가장 큰 수를 조회

print(find_max(nums))
```
