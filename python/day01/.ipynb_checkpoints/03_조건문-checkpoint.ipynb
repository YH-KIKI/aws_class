```python
# num가 음수인지 양수인지 0인지 판별하는 코드
num = 10
if num > 0 :
    print(f"{num}은 양수")
elif num < 0 :
    print(f"{num}은 음수")
else : 
    print(f"{num}은 0")
```

```python
# num이 홀수인지 짝수인지 판별하는 코드 

num = 11
match num % 2 :
    case 0 : 
        print(f"{num}은 짝수")

    case _ :
        print(f"{num}은 홀수")
```

```python
# 조건 선택 연산자를 대신하는 조건부 표현식
# 자바에서 조건 선택 연산자 => 조건식 ? 참 : 거짓 
# 파이썬에 조건부 표현식 => 참 if 조건식 else 거짓

num = 11
print (f"{num}은 {"짝수" if num % 2 == 0 else "홀수"}")
```
