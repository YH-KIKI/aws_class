```python
from tensorflow.keras import models, layers
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
# 붓꽃 데이터 생성


dataset = load_iris()
```

```python
# 독립, 종속 분리

X = dataset.data # 시각화 안할거라 [:, 2:4] 2개만 고르는 이 작업은안해도됨
y = dataset.target
```

```python
# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 정규화
nomalizer = layers.Normalization(axis=-1)
nomalizer.adapt(X_train)
```

```python
# 모델 설계
model = models.Sequential([
  nomalizer,
  layers.Dense(16, activation='relu'),
  layers.Dense(8, activation='relu'),
  layers.Dense(3, activation='softmax'), # 출력층
])
```

```python
# 모델 설정
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics = ['accuracy'])
```

```python
# 학습
model.fit(X_train, y_train, epochs=1000, batch_size=16, verbose=0)
```

```python
# 실제(100번지) 6.3, 3.3, 6, 2.5 -> 2
# 예측 6.4, 3.2, 6, 2.4 -> 2
# 예측할 새로운 붓꽃
new_iris = np.array([[6.4, 3.2, 6, 2.4]], dtype='float32')
prediction = model.predict(new_iris)
```

```python
predic_res = np.argmax(prediction)
print(f'예측 품종 : {dataset.target_names[predic_res]}')
```
