```python
from gensim.models import Word2Vec
from konlpy.tag import Okt
import numpy as np
import pandas as pd
```

```python
# 문장 불러오기
# data/csv/sentiment_data_long

df = pd.read_csv('../../data/csv/sentiment_data_long.csv')
sentences = df['sentence']
```

```python
okt = Okt()

# 학교에 갔다
# 학교에 가다(19일차 MLP, RNN에서 했던 방식)
# [(학교), (에), (가다)] 현재 예제

tokened_sentences = [okt.morphs(s, stem=True) for s in sentences]
```

```python
model = Word2Vec(
  sentences = tokened_sentences, # 학습 시킬 문장들
  vector_size = 100, # 벡터화 시킬 차원 크기
  window = 5,	# 5단어씩
  min_count = 1, # 최소 1번은 나와야 함
  sg = 1, # Skip-gram 적용
)
```

```python
before_word = '음악'
after_word = '울리다'

# 각 단어의 벡터를 가져옴
v_before = model.wv[before_word]
v_after = model.wv[after_word]
# 각 단어와 가까운 단어를 구하기 위해 평균을 냄
context_vector = (v_before + v_after) / 2.0
# 평균과 가장 유사한 단어를 가져옴
sim_words = model.wv.most_similar(positive = [context_vector], topn=5) # 여러개의 벡터를 넣을수 있기 때문에 리스트로

# 앞 단어와 뒷 단어는 제외
result = [word for word, _ in sim_words if word not in [before_word, after_word]] 	# _ (빈문자열) 활용해서 담아서 버림!

# 결과를 출력
result[0]
```
