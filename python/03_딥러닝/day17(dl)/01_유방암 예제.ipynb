```python
# 데이터 생성. 유방암 sklean

from sklearn.datasets import load_breast_cancer
from tensorflow.keras import models, layers
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


dataset = load_breast_cancer()
```

```python
# 독립, 종속 분리
X = dataset.data
y = dataset.target
```

```python
# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 정규화
nomalizer = layers.Normalization(axis=-1)
nomalizer.adapt(X_train)
```

```python
# 모델 설계. 입력층, 은닉층, 출력층

# 각 층마다 활성화 함수(입력층 제외)

model = models.Sequential([
  nomalizer,
  layers.Dense(64, activation='relu'),
  layers.Dense(32, activation='relu'),
  layers.Dense(16, activation='relu'),
  layers.Dense(1, activation='sigmoid'), # 출력층 / 분류가 2개라 sigmoid
])
```

```python
# 모델 설정 : 손실함수, 옵티마이저 , 평가지표
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics = ['accuracy'])
```

```python
# 학습
history = model.fit(X_train, y_train, epochs=1000, batch_size=50, verbose=0)
```

```python
# 평가
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'정확도 : {test_acc}')
```
