```python
import os
import cv2
import easyocr
```

```python
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'
```

```python
def recognize_image_save(image_path):
  img = cv2.imread(image_path)
  
  if img is None:
    print('이미지 로딩 실패!')
    return
  
	# 가로가 1000px 넘으면 1000px로 고정. 높이를 비율에 맞게 수정
  height, width = img.shape[:2]
  # print(height, width) 확인용
  if width > 1000:
    new_width = 1000
    # 1000 / width : 줄어든 너비 비율
    new_height = int(height * 1000 / width)
    img = cv2.resize(img, (new_width, new_height))
  try:
     reader = easyocr.Reader(['ko', 'en'], gpu = False)
     result = reader.readtext(img)
     for (_, text, prop) in result:
       print(f'내용 : {text} ({prop:.2f})')
  except Exception as e:
      print(f'오류 발생 : {e}')
    

```

```python
recognize_image_save('../../data/img/ocr1.jpg')
```
