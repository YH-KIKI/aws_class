```python
# 데이터 로딩 

import pandas as pd

df = pd.read_csv('../../data/csv/성적데이터.csv')
```

```python
# 독립(정보), 종속(답) 분리

import numpy as np

X, y = np.array(df['hour']).reshape(-1,1), np.where(df['score'] >= 60, 1, 0) # 60점 넘으면1, 아니면0 
```

```python
# 정규화 : 정보들의 값 범위가 제각각 -> 같은 범위로 맞추는 것 
# ex ) 학점 : 최대 4.5, 토익 최대 : 990

from tensorflow.keras import models, layers

normalizer = layers.Normalization(axis = -1)
normalizer.adapt(X)
```

```python
# 모델 설계 : 은닉층, 출력층을 설계
# 정규화를 모델 설계에 포함시킬 수 있음 

model = models.Sequential([
  normalizer,
  layers.Dense(16, activation='relu'),
  layers.Dense(32, activation='relu'),
  layers.Dense(1, activation='sigmoid')	# 출력층
])
```

```python
# 모델 설정 : 옵티마이저, 손실함수 등을 설정(평가지표)
# 콜백 함수 ( 성능 최적화를 위해 )
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
# 학습
history = model.fit(X, y, epochs=100, verbose=0, shuffle = True, validation_split=0.2)
```

```python
# 평가
eval = model.evaluate(X[16:], y[16:])
print(eval)
```

```python
# 예측

학습시간들 = [[5], [10], [1]]
예측결과들 = model.predict(np.array(학습시간들))

for i in range(len(학습시간들)):
  학습시간 = 학습시간들[i][0]
  예측결과 = 예측결과들[i][0]
  if 예측결과 >= 0.5:
    합격여부 = '합격'
  else:
    합격여부 = '불합격'


  print(f'학습시간 : {학습시간}, 예측 결과 : {예측결과 * 100:.0f}% 확률로 : {합격여부}')
```
