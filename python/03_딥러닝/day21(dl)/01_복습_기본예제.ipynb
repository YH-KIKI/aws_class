```python
# 데이터 로딩 

import pandas as pd

df = pd.read_csv('../../data/csv/성적데이터.csv')
```

```python
# 독립(정보), 종속(답) 분리

import numpy as np

X = np.array(df['hour']).reshape(-1,1)
y = df['score']
```

```python
# 훈련(학습), 테스트(확인) 분리
SIZE = len(X)	# 데이터가 몇개인지
TRAIN_SIZE = int(SIZE * 0.8) 	# 훈련을 80%만
X_train, X_test = X[ :TRAIN_SIZE], X[TRAIN_SIZE: ] # [처음부터 : TRAIN_SIZE 전까지 (훈련용)], [TRAIN_SIZE부터 : 끝까지 (테스트용)]
y_train, y_test = y[ :TRAIN_SIZE], y[TRAIN_SIZE: ] # [처음부터 : TRAIN_SIZE 전까지 (훈련용)], [TRAIN_SIZE부터 : 끝까지 (테스트용)]
```

```python
# 정규화 : 정보들의 값 범위가 제각각 -> 같은 범위로 맞추는 것 
# ex ) 학점 : 최대 4.5, 토익 최대 : 990

from tensorflow.keras import models, layers

normalizer = layers.Normalization(axis = -1)
normalizer.adapt(X_train)
```

```python
# 모델 설계 : 은닉층, 출력층을 설계
# 정규화를 모델 설계에 포함시킬 수 있음 

model = models.Sequential([
  normalizer,
  layers.Dense(16, activation='relu'),
  layers.Dense(32, activation='relu'),
  layers.Dense(1, activation='linear')	# 출력층
])
```

```python
# 모델 설정 : 옵티마이저, 손실함수 등을 설정(평가지표)
# 콜백 함수 ( 성능 최적화를 위해 )
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
```

```python
# 학습
history = model.fit(X_train, y_train, epochs=1000, verbose=0)
```

```python
# 평가
loss = history.history['loss'][99]
loss = np.sqrt(loss)
print(loss)
```

```python
# 예측

학습시간들 = [[5.0], [10]]
예측점수들 = model.predict(np.array(학습시간들))
for i in range(len(학습시간들)):
  학습시간 = 학습시간들[i][0]
  예측점수 = int(예측점수들[i][0]) 
  if 예측점수 < 0:
    예측점수 = 0
  elif 예측점수 > 100:
    예측점수 = 100
  print(f'학습시간 : {학습시간}, 예측 점수 : {예측점수}')
```
