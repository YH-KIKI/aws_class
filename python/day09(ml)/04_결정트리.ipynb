```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# 1. 데이터 로드
iris = load_iris() # 사이킷런에서 붓꽃데이터를 제공
```

```python
X = iris.data
y = iris.target
```

```python
# 훈련, 테스트 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# 학습

# 모델 생성 (나무의 깊이를 3으로 제한해서 너무 복잡하지 않게 설정)
model = DecisionTreeClassifier(max_depth=3, random_state=32)
model.fit(X_train, y_train)
```

```python
# 성능확인 
print(f'성능 : {model.score(X_test, y_test)}')
```

```python
# 4. 시각화 설정
plt.figure(figsize=(15, 10))
plot_tree(model, 
          feature_names=iris.feature_names, # 꽃잎/꽃받침 길이 이름
          class_names=iris.target_names,   # 품종 이름
          filled=True,                     # 색상 채우기
          rounded=True)                    # 박스 모서리 둥글게
plt.show()
```
