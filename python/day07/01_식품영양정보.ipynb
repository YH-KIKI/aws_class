```python
# 식품 영양 정보를 가져와서 csv파일로 저장
# 가져올 정보. 식품명, 탄수화물, 지방, 단백질, 나트륨, 총 칼로리

import requests
import pandas as pd
from bs4 import BeautifulSoup


serviceKey = '서비스키입력'

def food(service_key : str, pageNo : int = 1):
  url = 'https://apis.data.go.kr/1471000/FoodNtrCpntDbInfo02/getFoodNtrCpntDbInq02'

  params = {
    'serviceKey' : 'a44950f0fdef778f1a262d82dbdf8bdf3817ccdcaa90cbbd7a2c36f32c9f3451',
    'pageNo' : pageNo,
    'numOfRows' : 50,
	  'type' : 'xml',
  }

  try:
    response = requests.get(url, params=params)
    response.raise_for_status()

    soup = BeautifulSoup(response.text, "lxml-xml")
    rows = soup.find_all("item")

    result = []
    for row in rows:
      식품명 = row.find("FOOD_NM_KR").text
      탄수화물 = row.find("AMT_NUM6").text 
      지방 = row.find("AMT_NUM4").text
      단백질 = row.find("AMT_NUM3").text 
      나트륨 = row.find("AMT_NUM13").text 
      총칼로리 = row.find("AMT_NUM1").text 
      result.append({
        '식품명' : 식품명, 
        '탄수화물' : 탄수화물, 
        '지방' : 지방, 
        '단백질' : 단백질,
        '나트륨' : 나트륨,
        '총칼로리' : 총칼로리
      })

    return pd.DataFrame(result)

  except Exception as e:
    print(f'예외 발생 : {e}')




```

```python
total_df = pd.DataFrame()

for page in range(1, 6):
    df = food(serviceKey, page)
    total_df = pd.concat([total_df, df], ignore_index=True)

total_df.to_csv("식품_영양정보.csv", index=False, encoding="utf-8-sig")

print("저장 완료 🔥")
```
