```python
import random

def play_game():
  answer = random.randint(1, 10)

  while True:
    num = int(input("숫자를 맞춰보세요"))

    if num > answer:
      print("너무 커용")

    elif num < answer:
      print("너무 작아용")

    else:
      print("올 쫌 하는데요? 정답")
      break
play_game()
```

```python
balance = 0 # 잔액은  0

while True:
  print("1.입금 2.출금 3.잔액조회 4.종료")

  menu = int(input("메뉴 선택 : "))

  if menu == 1:
    입금액 = int(input("입금 금액 : "))
    balance += 입금액
    print(f"{입금액}원이 입금되었습니다")
  
  elif menu == 2:
    출금액 = int(input("출금 금액 : "))

    if balance >= 출금액:
      balance -= 출금액
      print(f"{출금액}원이 출금되었습니다")
    elif balance < 출금액:
      print("잔액이 부족합니다")
    
  elif menu == 3:
    print(f"{balance}원 남아있습니다.")
  
  elif menu == 4:
    print("프로그램 종료")
    break

```

```python
balance = 0

def deposit():
    global balance
    money = int(input("입금 금액: "))
    balance += money
    print("입금 완료")

def withdraw():
    global balance
    money = int(input("출금 금액: "))
    if balance >= money:
        balance -= money
        print(f"출금 완료")
    else:
        print("출금불가")

def check_balance():
    global balance
    print(f"잔액은 : {balance}")


while True:

    print("1.입금 2.출금 3.잔액조회 4.종료")

    menu = int(input("메뉴 선택: "))

    if menu == 1:
        deposit()

    elif menu == 2:
        withdraw()

    elif menu == 3:
        check_balance()

    elif menu == 4:
        break
```

```python
import random


while True: # 게임 전체 반복코드 
  answer = random.randint(1, 10)
  chance = 5

  while chance > 0:
    print(f"남은기회 {chance}번")

    num = int(input("숫자를 입력하세요"))

    if num > answer:
      print("넘 커용")

    elif num < answer:
      print("넘 작아용")
    
    else:
      print(f"{num} 정답")
      break

    chance -= 1

  if chance == 0:
    print(f"실패! 정답은 {answer} 메롱 약오르지")

  again = input("다시 할까요? (y/n)").lower()

  if again == "y":
    continue

  elif again == "n":
    print("게임종료")
    break

  else:
    print("잘못선택하셨습니다. 게임종료")
    break
```

```python

def max_num(a, b):
  if a > b:
    return a
  else:
    return b

print(max_num(3,7))
print(max_num(10,4))
```

```python

def max_three(a, b, c):
  if a > b and a > c:
      return a
  elif b > a and b > c:
     return b
  else:
    return c

print(max_three(3,7,5))
print(max_three(10,4,8))
```

```python
def my_abs(x):
  if x > 0:
    return x
  elif x < 0:
    return abs(x)
  else:
    return 0
print(my_abs(5))
print(my_abs(-3))
print(my_abs(0))
```
