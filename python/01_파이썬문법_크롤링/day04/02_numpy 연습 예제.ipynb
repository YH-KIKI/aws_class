```python
import numpy as np

# 홍길동 학생의 국어, 영어, 수학 성적을 콘솔로 입력 받아 numpy배열로 만들고 콘솔에 출력하세요

kor_score = input("국어 성적을 입력하세요")
eng_score = input("영어 성적을 입력하세요")
math_score = input("수학 성적을 입력하세요")

scores = []
scores.append(kor_score)
scores.append(eng_score)
scores.append(math_score)


scores_array = np.array(scores)


print(scores)



# 강사님 풀이

subjects = ["국어", "영어", "수학"]
list = []
# 과목수만큼 반복문
for subject in subjects:
  #성적을 입력
  num = input(f"{subject} 성적을 입력하세요 : ")

  #리스트에 추가
  list.append(int(num))

#리스트를 numpy배열로 변환
scores = np.array(list)
print(scores)
```

```python
# 학생의 평균을 출력하세요

print (f"평균 : {scores.mean()}")
print (f"평균 : {np.mean(scores)}")
```
