```python
#콘솔에서 입력
str = input("정수를 입력하세요 : ")
#자바에서 문자열 + 정수 가능? O
#파이썬에서 문자열 + 정수 가능? X
#print(str + 1) 에러발생 -> 문자열 + 정수를 했기 때문에
#int(문자열) -> 문자열을 정수로 변환
print(int(str) +1)
```

```python
# 정수를 입력받아 입력받은 정수가 홀수인지 짝수인지 출력하는 코드 -> if문

num = input("정수 입력 : ")
num = int(num) # 문자열을 정수로 변환 
# 아래코드처럼 한줄로 입력 OK
# num = int(input("정수를 입력하세요")) 

if num % 2 == 0:
  print (f"{num}은 짝수입니다")
else:
  print(f"{num}은 홀수입니다")

# 자바 조건선택연산자 : 조건식 ? 참값 : 거짓값
# 파이썬 조건부 표현식: 참값 if 조건식 else 거짓값
print(f"{num}은 {"짝수" if num % 2 == 0 else "홀수"}")
```

```python
# 1부터 10까지의 합을 출력하는 예제

min, max, result = (1, 10, 0)

for i in range(min, max + 1):
  result = result + i  # result += i

print(f"{min}에서 {max}까지의 합은 {result}입니다")

list = [i for i in range(min, max + 1)]
print(f"{min}에서 {max}까지의 합은 {sum(list)}입니다")
```

```python
# min부터 max까지의 수를 출력하는 함수를 만들고 1부터 10까지 출력하세요
# 함수명 : printNums
# 매개변수 : min, max
# 리턴값 : 없음

def printNums(min, max):
  # min이 max보다 크면 출력이 안됨 -> 출력되게 하기 위해 min, max를 바꿈
  if min > max:
    min, max = max, min
  for i in range(min, max + 1):
    print(i)

# printNums(1, 10) 
printNums(10, 1) #위에서 if로 min, max값이 바뀌면 바꾸라고 했기때문에 값은 동일
  
```
