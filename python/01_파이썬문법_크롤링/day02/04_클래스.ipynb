```python
# 2차원 좌표평면의 점을 나타내는 Point 클래스를 선언

# Point 클래스의 객체를 생성

class Point:
  # 클래스 변수
  name : str = "포인트 클래스"

  # 생성자
  def __init__(self, x:int = 0, y:int = 0):
    #인스턴스 변수
    self.__x = x
    self.__y = y

  # 함수
  def print(self):
    print(f"({self.__x}, {self.__y})")

  # x의 getter
  @property
  def x(self):
    return self.__x
  # x의 setter
  @x.setter
  def x(self, x):
    self.__x = x

  # y의 getter
  @property
  def y(self):
    return self.__y
  # y의 setter
  @y.setter
  def y(self, y):
    self.__y = y


print(Point.name)
p1 = Point(0,0)
p2 = Point(1,1)
p3 = Point()
p1.print()
p2.print()
p3. x = 100
p3. y = 100 # y는 아직 getter setter가 없어서 0
p3.print()

  
```

##### @Property (getter/setter)

자바에서 getter/setter는
이름 자체가 getxxx, setxxx

파이썬에서 getter/setter는 
변수처럼 사용

자바에서 setter사용
user.setUserName('홍길동')

파이썬에서 setter 사용
user.user_name('홍길동')

```python
# 클래스 변수 확인 예제

class A:
  # 클래스 변수 -> 모든 객체가 같은 값을 가짐
  name : str = "A클래스입니다"

  def __init__(self, age):
    #인스턴스 변수 -> 객체마다 다른 값을 가짐
    self.age = age

  def print_info(self):
    print(f"name : {A.name}")
    print(f"나이 : {self.age}")



a1 = A(10)
a1.print_info()
a2 = A(20)
a2.print_info()
A.name = "A클래스2입니다"
a1.age = 30
a1.print_info()
a2.print_info()
```

```python
# 은행 계좌 클래스
# 소유자(owner), 잔액(balance) => 인스턴스 변수
# 기능 : 출금(withdraw), 입금(deposit) => 함수

# 홍길동이 계좌 개설

# 100000(십만) 입금
# 50000(5만) 출금
# 70000(7만) 출금 -> 잔액 부족 알림

class BankAccount:
  name : str = "홍길동님 계좌입니다"

  # 계좌 개설
  def __init__(self, owner = "홍길동" , balance = 0):
    self.owner = owner
    self.balance = balance

  
  def 출금(self, 출금액:int):
    if self.balance >= 출금액:
      self.balance -= 출금액
      print(f"{출금액}이 출금되었습니다.")
      print(f"현재 잔액은 {self.balance}입니다.")

    else:
      print("현재 통장에 잔액이 부족합니다")
      print(f"현재 잔액은 {self.balance}입니다.")

  
  def 입금(self, 입금액:int):
    self.balance += 입금액
    print(f"{입금액}이 입금되었습니다.")
    print(f"현재 잔액은 {self.balance}입니다.")
    

# 홍길동이 계좌 개설 
account = BankAccount("홍길동")
# 100000(십만) 입금
account.입금(100000)
# 50000(5만) 출금
account.출금(50000)
# 70000(7만) 출금 -> 잔액 부족 알림
account.출금(70000)
 


  
```

```python
# 도서 대여 시스템에서 사용할 도서 클래스 선언 및 활용
# 도서 클래스
# 저자(author), 도서명(title), 대출 가능 여부(is_borrow)
# 기능 : 대출, 반납, 정보확인


class Book:
  # 생성자
  def __init__(self, title : str, author : str):
    self.__title = title
    self.__author = author
    self.__is_borrow = False  # 대출가능
    self.__borrower = ""

  # 대출 
  def borrow_book(self, user_name : str):
    # 대출 중이 아닐 때
    if not self.__is_borrow:
      print(f"{user_name}님이 {self.__title}을 대출했습니다.")
      self.__is_borrow = True
      self.__borrower = user_name
    
    # 대출중일 때
    else:
      print("이미 대출 중인 도서라 대출이 불가능 합니다.")
  

  # 반납
  def return_book(self):
    print(f"{self.__borrower}님이 {self.__title}을 반납했습니다. 해당 도서는 대출이 가능합니다.")

  #정보확인
  def show_info(self):
    print(f"제목 : {self.__title}")
    print(f"저자 : {self.__author}")
    print(f"대출 여부 : {"대출 중" if self.__is_borrow else "대출 가능"}")


book1 = Book("파이썬 정석", "홍길동")
book2 = Book("AI 원리", "고길동")

book1.borrow_book("회원1") # 회원1님이 파이썬 정석을 대출했습니다.
book1.show_info()


book1.borrow_book("회원2") # 이미 대출 중인 도서라 대출이 불가능 합니다.
book1. return_book() #회원1님이 파이썬 정석을 반납했습니다. 해당 도서는 대출이 가능합니다.

# 제목 : 파이썬 정석
# 저자 : 홍길동
# 대출 여부 : 대출 가능 # 이미 대출 중인 경우 대출 중
  
```
